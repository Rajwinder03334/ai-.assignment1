```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-5, 5, 100)

# Define the equations
y1 = 2*x + 1
y2 = 2*x + 2
y3 = 2*x + 3

# Plot the lines
plt.plot(x, y1, color='blue', linestyle='-', label='y = 2x + 1')
plt.plot(x, y2, color='red', linestyle='--', label='y = 2x + 2')
plt.plot(x, y3, color='green', linestyle='-.', label='y = 2x + 3')

# Add title and labels
plt.title('Graphs of Lines y = 2x + 1, y = 2x + 2, and y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define matrix A
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

# Calculate the inverse of matrix A
A_inv = np.linalg.inv(A)

# Calculate the product of A and its inverse
product_AA_inv = np.dot(A, A_inv)

# Calculate the product of the inverse of A and A
product_A_inv_A = np.dot(A_inv, A)

# Print the results
print("Matrix A:")
print(A)
print("\nInverse of matrix A:")
print(A_inv)
print("\nProduct of A and its inverse (A * A^-1):")
print(product_AA_inv)
print("\nProduct of the inverse of A and A (A^-1 * A):")
print(product_A_inv_A)
```
